# Loading Beautiful Soup and the source URL

```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
# URL of page to be scraped
url = 'https://redplanetscience.com/'

# Retrieve page with the requests module
response = requests.get(url)

response.text
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

# JPL Mars Space Images - Featured Image

```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Chrome Browser

browser = Browser('chrome', executable_path=ChromeDriverManager().install(), headless=False)
```

```python
# URL of page to be scraped
url = 'https://spaceimages-mars.com'

browser.visit(url)
```

```python
html = browser.html
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')

image_url = soup.find_all('a', class_= "showimg")
for img in image_url:
    print(img)
```

```python
href = img['href']
href
```

```python
featured_image_url = ('https://spaceimages-mars.com/' + href)
featured_image_url
```
